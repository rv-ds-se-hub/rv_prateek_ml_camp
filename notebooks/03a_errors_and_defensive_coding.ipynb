## EAFP - Easier to Ask For Forgiveness Than Permission

```python
# EAFP
val = "-42"

try:
    val = int(val)
except ValueError:
    val = 0

print(val)
```

```python
# EAFP
greetings = ['morning', 'afternoon', 'evening']
index = 3

with open("greeting.txt", 'w', encoding='utf-8') as hello:
    try:
        hello.write("Good\t")
        hello.write(greetings[index])
    except IndexError:
        print("Error. Big Index.")
```

```python
# LBYL
if index >= len(greetings):
    print("Error")
else:
    with open("greeting.txt", "w") as hello:
        hello.write(f"Good, {greetings[index]}")
```

```python
# EAFP (smart)

with open("greetings.txt", "w", encoding="utf-8") as hello:
    try:
        greeting = greetings[index]
    except IndexError:
        print("Error...")
    else:
        hello.write(f"Hello, {greeting}")
    finally:
        print("This will run regardless of what happens.")
```
